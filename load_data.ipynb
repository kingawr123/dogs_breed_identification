```python
import numpy as np
import pandas as pd
import cv2
import os
```

```python
IMAGE_SIZE = (331, 331)
IMAGE_FULL_SIZE = (331, 331, 3)

train_images_folder_path = "train"
test_images_folder_path = "test"

labels_df = pd.read_csv("labels.csv")
labels_df.head(10)
```

```python
train_images = []
train_labels = []

for i, (image_name, breed) in enumerate(labels_df[['id', 'breed']].values):
    img_path = os.path.join(train_images_folder_path, image_name + '.jpg')
    img = cv2.imread(img_path)
    resized = cv2.resize(img, IMAGE_SIZE, interpolation=cv2.INTER_AREA) / 255.0
    train_images.append(resized)
    train_labels.append(breed)
    

print(len(train_images))
print(len(train_labels))

assert len(train_images) == len(train_labels)

print("saving data ...")

# np.save("train_images.npy", train_images)
np.save("train_labels.npy", train_labels)

print("done")
```
